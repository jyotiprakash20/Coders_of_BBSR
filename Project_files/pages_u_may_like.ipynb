```python
import json
```

```python
def load_data(filename):
    with open(filename, 'r') as f:
        return json.load(f)
    
```

```python
#Find pages a user may like based on their liked pages
def pages_you_might_like(user_id, data):

    #Dictonary to store user interaction with pages
    user_pages={}

    for user in data['users']:
        user_pages[user['id']]= set(user['liked_pages'])
    
    #if the user is not found return empty list
    if user_id not in user_pages:
        return []
    
    user_liked_pages=user_pages[user_id]
    page_recommendations={}

    for other_user, pages in user_pages.items():
        if other_user !=user_id:
            shared_pages=user_liked_pages.intersection(pages)
        for page in pages:
            if page not in user_liked_pages:
                page_recommendations[page]=page_recommendations.get(page,0)+len(shared_pages)
            
    sorted_pages=sorted(page_recommendations.items(), key=lambda x: x[1], reverse=True)
    return [(page_id,score) for page_id, score in sorted_pages]

data=load_data('Massive_data.json')
user_id=1
recommended_pages=pages_you_might_like(user_id, data)
print(f"Pages recommended for user {user_id}: {recommended_pages}")
```
